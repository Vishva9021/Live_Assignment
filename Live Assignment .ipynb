```python
#1.10. Write a program to find the sum of all the positive numbers entered by the user. As soon as the user enters a negative number, stop taking in any further input from the user and display the sum .

def sum_of_positive_numbers():
    total_sum = 0
    while True:
        user_input = input("Please enter a number: ")
        try:
            number = float(user_input)
            if number < 0:
                print("You entered a negative number. Stopping input.")
                break
            elif number > 0:
                total_sum += number
            else:
                print("You entered zero. It won't be added to the sum.")
        except ValueError:
            print("That's not a valid number. Please try again.")
    
    print(f"The sum of all positive numbers entered is: {total_sum}")

if __name__ == "__main__":
    sum_of_positive_numbers()
```

```python
#1.11. Write a program to find prime numbers between 2 to 100 using nested for loops.
def find_prime_numbers():
    prime_numbers = []
    
    for possible_prime in range(2, 101):
        is_prime = True
        
        for num in range(2, int(possible_prime ** 0.5) + 1):
            if possible_prime % num == 0:
                is_prime = False
                break
        
        if is_prime:
            prime_numbers.append(possible_prime)
    
    print("Prime numbers between 2 and 100 are:", prime_numbers)

if __name__ == "__main__":
    find_prime_numbers()
```

```python
""""4.Create a simple banking system using object-oriented concepts in Python. Design classes representing
different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw,
and balance inquiry. Utilize inheritance to manage different account types efficiently."""

# Base class for bank accounts
class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.account_number = account_number
        self.balance = initial_balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount} into account {self.account_number}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount} from account {self.account_number}. New balance: ${self.balance}")

    def balance_inquiry(self):
        print(f"Account {self.account_number} balance: ${self.balance}")


# Subclass for SavingsAccount
class SavingsAccount(BankAccount):
    def __init__(self, account_number, initial_balance, interest_rate):
        super().__init__(account_number, initial_balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate / 100
        self.balance += interest
        print(f"Added ${interest} interest to account {self.account_number}. New balance: ${self.balance}")


# Subclass for CheckingAccount
class CheckingAccount(BankAccount):
    def __init__(self, account_number, initial_balance, overdraft_limit):
        super().__init__(account_number, initial_balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > self.balance + self.overdraft_limit:
            print("Transaction exceeds overdraft limit.")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount} from account {self.account_number}. New balance: ${self.balance}")


# Testing the banking system
def main():
    savings_account = SavingsAccount("SAV123", 1000, 5)
    print("Savings Account Transactions:")
    savings_account.deposit(500)
    savings_account.withdraw(200)
    savings_account.balance_inquiry()
    savings_account.add_interest()

    checking_account = CheckingAccount("CHK456", 500, 1000)
    print("\nChecking Account Transactions:")
    checking_account.deposit(200)
    checking_account.withdraw(700)
    checking_account.balance_inquiry()


if __name__ == "__main__":
    main()
```

```python
#6.Write a code for Restaurant Management System Using OOPS

class MenuItem:
    _id_counter = 1  # Private class variable to generate unique IDs

    def __init__(self, name, description, price, category):
        self.__id = MenuItem._id_counter  # Private instance variable for unique ID
        MenuItem._id_counter += 1
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    def get_id(self):
        return self.__id  # Getter method to access private ID

    def update_item(self, name=None, description=None, price=None, category=None):
        if name:
            self.name = name
        if description:
            self.description = description
        if price:
            self.price = price
        if category:
            self.category = category

    def __str__(self):
        return f"ID: {self.__id}, Name: {self.name}, Description: {self.description}, Price: {self.price}, Category: {self.category}"


class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, calories):
        super().__init__(name, description, price, category)
        self.calories = calories

    def update_item(self, name=None, description=None, price=None, category=None, calories=None):
        super().update_item(name, description, price, category)
        if calories:
            self.calories = calories

    def __str__(self):
        return f"{super().__str__()}, Calories: {self.calories}"


class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, volume):
        super().__init__(name, description, price, category)
        self.volume = volume

    def update_item(self, name=None, description=None, price=None, category=None, volume=None):
        super().update_item(name, description, price, category)
        if volume:
            self.volume = volume

    def __str__(self):
        return f"{super().__str__()}, Volume: {self.volume}"


# Testing the Restaurant Management System
menu_items = []

def add_menu_item(item):
    menu_items.append(item)

def remove_menu_item(item_id):
    for item in menu_items:
        if item.get_id() == item_id:
            menu_items.remove(item)
            print(f"Menu item {item_id} removed.")
            return
    print(f"Menu item {item_id} not found.")

def update_menu_item(item_id, **kwargs):
    for item in menu_items:
        if item.get_id() == item_id:
            item.update_item(**kwargs)
            print(f"Menu item {item_id} updated.")
            return
    print(f"Menu item {item_id} not found.")

def display_menu():
    for item in menu_items:
        print(item)

# Create menu items
item1 = FoodItem("Burger", "Beef burger with cheese", 10.99, "Main Course", 500)
item2 = BeverageItem("Coca-Cola", "Soft drink", 2.99, "Drinks", 330)

# Add menu items
add_menu_item(item1)
add_menu_item(item2)

# Display menu
print("Menu:")
display_menu()

# Update menu item
update_menu_item(1, price=11.99)

# Remove menu item
remove_menu_item(2)

# Display updated menu
print("\nUpdated Menu:")
display_menu()
```

```python
#7.Write a code for Hotel Management System using OOPS 

class Room:
    _id_counter = 1  # Private class variable to generate unique IDs

    def __init__(self, room_number, room_type, rate):
        self.__id = Room._id_counter  # Private instance variable for unique ID
        Room._id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = True  # Private instance variable for availability
        self.guest = None  # Instance variable to store guest information

    def get_id(self):
        return self.__id  # Getter method to access private ID

    def is_available(self):
        return self.__availability  # Getter method to access private availability

    def book_room(self, guest):
        if self.__availability:
            self.__availability = False
            self.guest = guest
            print(f"Room {self.room_number} booked for {guest}.")
        else:
            print(f"Room {self.room_number} is not available.")

    def check_in(self, guest):
        if not self.__availability:
            print(f"Guest {guest} checked in to room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked for {guest}.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            self.guest = None
            print(f"Guest checked out of room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is already available.")


class SuiteRoom(Room):
    def __init__(self, room_number, rate, extra_bed):
        super().__init__(room_number, "Suite", rate)
        self.extra_bed = extra_bed

    def __str__(self):
        return f"Suite Room {self.room_number}, Rate: {self.rate}, Extra Bed: {self.extra_bed}"


class StandardRoom(Room):
    def __init__(self, room_number, rate, view):
        super().__init__(room_number, "Standard", rate)
        self.view = view

    def __str__(self):
        return f"Standard Room {self.room_number}, Rate: {self.rate}, View: {self.view}"


# Testing the Hotel Management System
rooms = []

def add_room(room):
    rooms.append(room)

def book_room(room_number, guest):
    for room in rooms:
        if room.room_number == room_number:
            room.book_room(guest)
            return
    print(f"Room {room_number} not found.")

def check_in(room_number, guest):
    for room in rooms:
        if room.room_number == room_number:
            room.check_in(guest)
            return
    print(f"Room {room_number} not found.")

def check_out(room_number):
    for room in rooms:
        if room.room_number == room_number:
            room.check_out()
            return
    print(f"Room {room_number} not found.")

def display_rooms():
    for room in rooms:
        print(room)

# Create rooms
room1 = SuiteRoom(101, 200, True)
room2 = StandardRoom(202, 150, "City View")

# Add rooms
add_room(room1)
add_room(room2)

# Display rooms
print("Rooms:")
display_rooms()

# Book a room
book_room(101, "John Doe")

# Check in
check_in(101, "John Doe")

# Check out
check_out(101)

# Display updated rooms
print("\nUpdated Rooms:")
display_rooms()
```

```python
#9.Write a code for Event Management System using OOPS

class Event:
    _id_counter = 1  # Private class variable to generate unique IDs

    def __init__(self, name, date, time, location):
        self.__id = Event._id_counter  # Private instance variable for unique ID
        Event._id_counter += 1
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []  # Private instance variable for list of attendees

    def get_id(self):
        return self.__id  # Getter method to access private ID

    def add_attendee(self, attendee):
        self.__attendees.append(attendee)
        print(f"Added {attendee} to {self.name} event.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Removed {attendee} from {self.name} event.")
        else:
            print(f"{attendee} is not attending {self.name} event.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def __str__(self):
        return f"Event {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}"


class PrivateEvent(Event):
    def __init__(self, name, date, time, location, password):
        super().__init__(name, date, time, location)
        self.password = password

    def add_attendee(self, attendee, password):
        if password == self.password:
            super().add_attendee(attendee)
        else:
            print("Invalid password. Cannot add attendee.")

    def __str__(self):
        return f"Private Event {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}, Password: {self.password}"


class PublicEvent(Event):
    def __init__(self, name, date, time, location, capacity):
        super().__init__(name, date, time, location)
        self.capacity = capacity

    def add_attendee(self, attendee):
        if len(self.__attendees) < self.capacity:
            super().add_attendee(attendee)
        else:
            print("Event is fully booked. Cannot add attendee.")

    def __str__(self):
        return f"Public Event {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}, Capacity: {self.capacity}"


# Testing the Event Management System
events = []

def add_event(event):
    events.append(event)

def display_events():
    for event in events:
        print(event)

# Create events
event1 = PrivateEvent("Wedding", "2023-02-14", "10:00 AM", "Hotel", "love123")
event2 = PublicEvent("Concert", "2023-03-20", "7:00 PM", "Stadium", 5000)

# Add events
add_event(event1)
add_event(event2)

# Display events
print("Events:")
display_events()

# Add attendees
event1.add_attendee("John Doe", "love123")
event2.add_attendee("Jane Doe")

# Display updated events
print("\nUpdated Events:")
display_events()

# Get total attendees
print(f"\nTotal attendees for {event1.name}: {event1.get_total_attendees()}")
print(f"Total attendees for {event2.name}: {event2.get_total_attendees()}")
```

```python
#10.Write a code for Airline Reservation System using OOPS


class Flight:
    _id_counter = 1  # Private class variable to generate unique IDs

    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats):
        self.__id = Flight._id_counter  # Private instance variable for unique ID
        Flight._id_counter += 1
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = available_seats  # Private instance variable for available seats

    def get_id(self):
        return self.__id  # Getter method to access private ID

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked on flight {self.flight_number}. Remaining seats: {self.__available_seats}")
        else:
            print(f"Flight {self.flight_number} is fully booked.")

    def cancel_reservation(self):
        if self.__available_seats < self.__available_seats:
            self.__available_seats += 1
            print(f"Reservation cancelled on flight {self.flight_number}. Remaining seats: {self.__available_seats}")
        else:
            print(f"No reservations to cancel on flight {self.flight_number}.")

    def get_remaining_seats(self):
        return self.__available_seats

    def __str__(self):
        return f"Flight {self.flight_number}, Departure: {self.departure_airport} at {self.departure_time}, Arrival: {self.arrival_airport} at {self.arrival_time}, Available Seats: {self.__available_seats}"


class Passenger:
    def __init__(self, name, email):
        self.name = name
        self.email = email

    def __str__(self):
        return f"Passenger {self.name}, Email: {self.email}"


# Testing the Airline Reservation System
flights = []

def add_flight(flight):
    flights.append(flight)

def display_flights():
    for flight in flights:
        print(flight)

def book_seat(flight_number, passenger):
    for flight in flights:
        if flight.flight_number == flight_number:
            flight.book_seat()
            print(f"Seat booked for {passenger} on flight {flight_number}.")
            return
    print(f"Flight {flight_number} not found.")

def cancel_reservation(flight_number, passenger):
    for flight in flights:
        if flight.flight_number == flight_number:
            flight.cancel_reservation()
            print(f"Reservation cancelled for {passenger} on flight {flight_number}.")
            return
    print(f"Flight {flight_number} not found.")

# Create flights
flight1 = Flight("AA101", "JFK", "LAX", "08:00 AM", "11:00 AM", 200)
flight2 = Flight("UA202", "SFO", "ORD", "12:00 PM", "05:00 PM", 150)

# Add flights
add_flight(flight1)
add_flight(flight2)

# Display flights
print("Flights:")
display_flights()

# Book seats
passenger1 = Passenger("John Doe", "johndoe@example.com")
passenger2 = Passenger("Jane Doe", "janedoe@example.com")
book_seat("AA101", passenger1)
book_seat("UA202", passenger2)

# Cancel reservations
cancel_reservation("AA101", passenger1)
cancel_reservation("UA202", passenger2)

# Display updated flights
print("\nUpdated Flights:")
display_flights()
```

```python
"""15. Write a Python module named file_operations.py with functions for reading, writing, and appending
data to a file."""
# file_operations.py

def read_file(file_path):
    """Reads the contents of a file and returns it as a string"""
    try:
        with open(file_path, 'r') as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

def write_file(file_path, data):
    """Writes data to a file, overwriting any existing contents"""
    try:
        with open(file_path, 'w') as file:
            file.write(data)
    except IOError:
        print(f"Error: Unable to write to file '{file_path}'.")

def append_file(file_path, data):
    """Appends data to the end of a file"""
    try:
        with open(file_path, 'a') as file:
            file.write(data)
    except IOError:
        print(f"Error: Unable to append to file '{file_path}'.")

def read_lines(file_path):
    """Reads the contents of a file and returns it as a list of lines"""
    try:
        with open(file_path, 'r') as file:
            return file.readlines()
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

def write_lines(file_path, lines):
    """Writes a list of lines to a file, overwriting any existing contents"""
    try:
        with open(file_path, 'w') as file:
            file.writelines(lines)
    except IOError:
        print(f"Error: Unable to write to file '{file_path}'.")

def append_lines(file_path, lines):
    """Appends a list of lines to the end of a file"""
    try:
        with open(file_path, 'a') as file:
            file.writelines(lines)
    except IOError:
        print(f"Error: Unable to append to file '{file_path}'.")
```

```python
"""18. Create a Python script that reads a text file named "expenses.txt" and calculates the total amount spent
on various expenses listed in the file."""
# calculate_expenses.py

def calculate_total_expenses():
    total_expenses = 0.0

    try:
        with open("expenses.txt", "r") as file:
            for line in file:
                expense_info = line.strip().split(",")
                expense_amount = float(expense_info[1])
                total_expenses += expense_amount
                print(f"Expense: {expense_info[0]}, Amount: {expense_amount:.2f}")
    except FileNotFoundError:
        print("Error: File 'expenses.txt' not found.")
    except ValueError:
        print("Error: Invalid expense amount in file.")

    print(f"Total Expenses: {total_expenses:.2f}")

if __name__ == "__main__":
    calculate_total_expenses()
```

```python
"""19. Create a Python program that reads a text file named "paragraph.txt" and counts the occurrences of
each word in the paragraph, displaying the results in alphabetical order."""
# word_counter.py

def count_words():
    word_counts = {}

    try:
        with open("paragraph.txt", "r") as file:
            paragraph = file.read()
            words = paragraph.lower().split()
            for word in words:
                word = word.strip(".,;:!?'\"")
                if word not in word_counts:
                    word_counts[word] = 1
                else:
                    word_counts[word] += 1
    except FileNotFoundError:
        print("Error: File 'paragraph.txt' not found.")

    for word, count in sorted(word_counts.items()):
        print(f"{word}: {count}")

if __name__ == "__main__":
    count_words()
```
